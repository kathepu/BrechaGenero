```python
import pandas as pd

df = pd.read_csv("../data/casen_sintetico.csv")
df.head()
```

```python
df["ocupado"] = df["situacion_laboral"] == "Ocupado"
df[["situacion_laboral", "ocupado"]].head()
```

```python
df.groupby("situacion_laboral")["ingreso_mensual"].apply(lambda x: x.isna().sum())
```

```python
df.to_csv("../data/casen_limpio.csv", index=False)
print("Archivo guardado en data/casen_limpio.csv")
```
